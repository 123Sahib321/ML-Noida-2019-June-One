```python
theta = [2, 2]
```

```python
X = np.linspace(10, 100, 10)
```

```python
noise = np.random.randint(-20, 20, len(X))
```

```python
y = hypothesis(theta, X) + noise
```

```python
plt.figure()
plt.scatter(X, y)
plt.ylim(0, 300)
plt.show()
```

```python
class LinearRegression:
    
    def __init__(self, lr=.0001, itrs=10000):
        self.lr = lr
        self.itrs = itrs
    
    def fit(self, X, y):
        self.X = X
        self.y = y
        
        self.theta = np.zeros([2], dtype=float)
        for i in range(self.itrs):
            self.gradient_dec()
        
    def gradient_dec(self):
        yh = LinearRegression.hypothesis(self.theta, self.X)
        
#         print(yh)
        
        change_slop = np.sum((yh-self.y) * X) / len(X) 
        change_inter = np.sum((yh-self.y)) / len(X)
        
#         print((yh-self.y) * X)
#         print((yh-self.y))
        
        change_slop = change_slop * self.lr
        change_inter = change_inter * self.lr
        
        self.theta[0] =  self.theta[0] - change_inter
        self.theta[1] = self.theta[1] - change_slop
        
        print(self.theta)
        
    def predict(self, X):
        return LinearRegression.hypothesis(self.theta, X)
    
    
    
    @classmethod
    def hypothesis(cls, theta, X):
        return X * theta[1] + theta[0]
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
yp = model.predict(X)
```
